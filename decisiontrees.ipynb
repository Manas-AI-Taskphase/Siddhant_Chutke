**DECISION TREES**

```python
import math
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('threats.csv')
df
```

```python
df[df.isnull().any(axis=1)].head()
```

```python
df=df.dropna()
#null row drop
```

```python
y = df['Threat'].copy()
```

```python
x = df.drop(['Threat'], axis = 1)
x
# storing that column in a different variable, for assessment
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state = 311) #defining train and test vars
```

train and test data done


```python
totchg_classifier = DecisionTreeClassifier(max_leaf_nodes=7, random_state=12)
totchg_classifier.fit(x_train, y_train) #decisiontreeclassifier
```

```python
y_prediction = totchg_classifier.predict(x_test)
accuracy_score(y_test, y_prediction)*100
```

```python


confusion_matrix(y_test, y_prediction)
```
